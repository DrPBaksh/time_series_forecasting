## Identifying the Multiple Seasonalities in the time series

```python
from statsmodels.graphics.tsaplots import plot_acf
acf = plot_acf(X_cleaned['total load actual'])
plt.grid('minor')
plt.show()
```

```python
day = X_cleaned['total load actual'].groupby(pd.Grouper(freq = 'D')).mean()
acf = plot_acf(day)
plt.show()
```

```python
week = X_cleaned['total load actual'].groupby(pd.Grouper(freq = 'W')).mean()
acf = plot_acf(week)
plt.show()
```

```python
month = X_cleaned['total load actual'].groupby(pd.Grouper(freq = 'M')).mean()
acf = plot_acf(month)
plt.show()
```

```python

import statsmodels.api as sm


# Fit an ETS model with two seasonal periods
ets_model = sm.tsa.ETSModel(X['total load actual'].head(1000) ,error='add', trend=None, seasonal=None, seasonal_periods=[24, int(24*7), int(365*24)]).fit()

# Generate a 30-step-ahead forecast
forecast = ets_model.forecast(steps=30)

# Plot the residuals of the fitted model
residuals = ets_model.resid
plt.plot(residuals)
plt.title('Residuals of ETS model')
plt.show()

forecast
```
